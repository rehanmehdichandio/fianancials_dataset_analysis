# Financials Dataset EDA and Visualization

```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file
df = pd.read_csv('/content/Financials.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns=[' Discounts '])
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python

# List of columns to clean
columns_to_clean = [
    ' Units Sold ',
    ' Manufacturing Price ',
    ' Sale Price ',
    ' Gross Sales ',
    '  Sales ',
    ' COGS ',
    ' Profit '
]

for col in columns_to_clean:
    df[col] = df[col].replace(r'[\$,]', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.head()
```

```python
category_type = ['Segment', 'Country', ' Product ', ' Discount Band ']
for col in category_type:
  df[col] = df[col].astype('category')
```
